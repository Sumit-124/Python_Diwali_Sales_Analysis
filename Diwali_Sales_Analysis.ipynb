# Exploratory Data Analysis

## Gender

```python
plt.style.available
plt.style.use('tableau-colorblind10')
```

```python
df.columns
```

```python
# count no. of males and females.
count_gender = df.groupby('Gender').count()
count_gender['User_ID']
```

```python
# Plot count of user_Id vs Gender.
count_gender['User_ID'].plot(kind='bar', xlabel='Gender', ylabel='Count')
```

```python
# yha pe phele humne gender ko grouo by kara, the sum function lagaya to find the total amount of products purchased by males and females.
sum_gender = df.groupby('Gender').sum()
sum_gender['Amount']
```

```python
sum_gender['Amount'].plot(kind='bar')
```

##### From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men.

## State

```python
# Total no. of orders from top 10 states
df
```

```python
count_States = df.groupby('State', as_index= False ).count().sort_values(by='Orders', ascending=False).head(10)
count_States
```

```python
count_States.set_index('State', inplace=True)
count_States['Orders'].plot(kind='bar')
```

##### As we can in the above graph that the max no. of orders came from Uttar Pradesh

```python
# Now we want to see how much amount of order came from the top 10 states.
sum_State = df.groupby('State', as_index=False).sum().sort_values(by='Amount', ascending=False).head(10)
sum_State
```

```python
sum_State.set_index('State', inplace=True)
sum_State['Amount'].plot(kind='bar')
```

##### As we can say in the above graph that most of the orders and total sales/amount are from uttar pradesh, maharashtra and karnataka respectively.

## Marital Status

```python
df
```

```python
# now we want see ki married wale kitne hai jitno ne order kiye hai and unmmaried kitne hai jinhone order kare hai
# means We want to count the no. of married and unmarried who ordered.
count_marital_status = df.groupby('Marital_Status').count()
count_marital_status
```

## Occupation

```python
df
```

```python
# Now we want see kitne kog vaise hai jinhone jyada order kre hai and same occupation se belong karte hai
count_occupation = df.groupby('Occupation', as_index=False).count().sort_values(by='Orders', ascending=False)
count_occupation
count_occupation.set_index('Occupation', inplace=True)
count_occupation['Orders'].plot(kind='bar', xlabel='occupation', ylabel='No. of orders')
```

```python
sum_occupation = df.groupby('Occupation', as_index=False).sum().sort_values(by='Amount', ascending=False)
sum_occupation
sum_occupation.set_index('Occupation', inplace=True)
sum_occupation['Amount'].plot(kind='bar', xlabel='Occupation',ylabel='Amount Spent')
```

##### From the above graph we can see that most of the buyers are working in IT Sector, Healthcare and Aviation industry.

## Product Category

```python
df.columns
```

```python
df
```

```python
# now we want to check category wise ki kis category ke product ke orders sabse jyada aaye hai like books ke orders jyada aaye hai?, 
# beauty products ke orders jyada aaye hai? or what?
# No. of orders VS Product_Category
# top 10 most sold products konse hai?
count_product_category = df.groupby('Product_Category', as_index=False).count().sort_values(by='Orders', ascending=False).head(10)
count_product_category
count_product_category.set_index('Product_Category', inplace=True)
count_product_category['Orders'].plot(kind='bar', xlabel='Product_Category', ylabel='no. of orders sold')
```

```python
sum_product_category = df.groupby('Product_Category', as_index=False).sum().sort_values(by='Amount', ascending=False).head(10)
sum_product_category
sum_product_category.set_index('Product_Category', inplace=True)
sum_product_category['Amount'].plot(kind='bar', xlabel='Product_Category', ylabel='Amount', figsize=(15,5))
```

##### From the above graph we can see that most of the sold products are from Food, Clothing & Apparel and Electronics & Gadgets category
